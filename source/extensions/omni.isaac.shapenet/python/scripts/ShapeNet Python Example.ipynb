```python
from threading import Thread
import asyncio
import time

async def do_some_work(x):
    print("Start awating work " + str(x))
    await asyncio.sleep(x)
    print("Finished awating work ", x)

loop = asyncio.get_event_loop()
print("run_until_complete 5")
loop.run_until_complete(do_some_work(5))
print("complete 5")

tasks = [asyncio.ensure_future(do_some_work(6)),
         asyncio.ensure_future(do_some_work(5))]
print("run_until_complete [6,5]")
loop.run_until_complete(asyncio.gather(*tasks))
print("complete [6,5]")

def start_loop(loop):
    asyncio.set_event_loop(loop)
    loop.run_forever()

new_loop = asyncio.new_event_loop()
t = Thread(target=start_loop, args=(new_loop,))
t.start()

def more_work(x):
  print("More work %s" % x)
  time.sleep(x)
  print("Finished more work %s" % x)

#new_loop.call_soon_threadsafe(more_work, 6)
#new_loop.call_soon_threadsafe(more_work, 3)

new_loop.call_soon_threadsafe(more_work, 20)
asyncio.run_coroutine_threadsafe(do_some_work(5), new_loop)
asyncio.run_coroutine_threadsafe(do_some_work(10), new_loop)

new_loop.stop()
new_loop.close()
```

```python
import asyncio
import aiohttp

async def fetch(url):
    """Perform an HTTP GET to the URL and print the response"""
    response = await aiohttp.request('GET', url)
    return await response.text()

# Get a reference to the event loop
loop = asyncio.get_event_loop()
# Create the batch of requests we wish to execute
requests = [asyncio.ensure_future(fetch("https://github.com")),
            asyncio.ensure_future(fetch("https://google.com"))]
# Run the batch
responses = loop.run_until_complete(asyncio.gather(*requests))
# Examine responses
for resp in responses:
  print(resp)
```

Stopping the loop

```python
try:
    loop.run_forever()
except KeyboardInterrupt:
    # Canceling pending tasks and stopping the loop
    asyncio.gather(*asyncio.Task.all_tasks()).cancel()
    # Stopping the loop
    loop.stop()    
    # Received Ctrl+C
    loop.close()
```
