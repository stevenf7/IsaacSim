```python
from isaacsim import SimulationApp

simulation_app = SimulationApp()

# Any Omniverse level imports must occur after the `SimulationApp` class is instantiated (because APIs are provided
# by the extension/runtime plugin system, it must be loaded before they will be available to import).
import isaacsim.core.experimental.utils.app as app_utils
import isaacsim.core.experimental.utils.stage as stage_utils
from isaacsim.core.experimental.objects import Cube
from isaacsim.core.experimental.prims import GeomPrim, RigidPrim
```

```python
# Create a new stage
stage_utils.create_new_stage(template="gridroom")
# Create a cube
fancy_cube_shape = Cube(
    paths="/World/random_cube",
    positions=[0, 0, 1.0],
    sizes=1.0,
    scales=[0.5015, 0.505, 0.5015],
    colors=[0.0, 0.0, 1.0],
)
# Apply physics (RigidPrim) and collision (GeomPrim) to the cube
fancy_cube = RigidPrim(paths=fancy_cube_shape.paths)
fancy_cube_geometry = GeomPrim(paths=fancy_cube_shape.paths, apply_collision_apis=True)
simulation_app.update()
```

```python
app_utils.play()
simulation_app.update()  # Allow physics to initialize

for i in range(100):
    positions, orientations = fancy_cube.get_world_poses()
    linear_velocities, angular_velocities = fancy_cube.get_velocities()
    print("Cube position is : " + str(positions))
    print("Cube's orientation is : " + str(orientations))
    print("Cube's linear velocity is : " + str(linear_velocities))
    # Step a physics and rendering update
    simulation_app.update()
```

```python
# Cleanup application
simulation_app.close()
```
